```python
plt.plot(timestep,average_dqn, label='DQN')
plt.errorbar(timestep, average_dqn, yerr=std_deviation_dqn, fmt='o', capsize=5, label='DQN Average Returns and Average Std Devs')
plt.plot(timestep, average_ddqn, label='DDQN')
plt.errorbar(timestep, average_ddqn, yerr=std_deviation_ddqn, fmt='o', capsize=5, label='DDQN Average Returns and Average Std Devs')
plt.grid(True)
plt.xticks(timestep,rotation='vertical')

plt.legend()
plt.xlabel('Number of Iterations')
plt.ylabel('Train Average Returns')
plt.title('DQN and DDQN Average Returns vs Number of Itterations')
```
