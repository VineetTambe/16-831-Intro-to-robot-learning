```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Ant
avg_ret = "/home/vineet/Downloads/irl_hw1_data/q2_dagger_humanoid_20_Humanoid-v2_avg_ret.csv"  # Replace with the path to your first CSV file
df1 = pd.read_csv(avg_ret)

# Load the second CSV file
avg_std_dev = "/home/vineet/Downloads/irl_hw1_data/q2_dagger_humanoid_20_Humanoid-v2_std_dev.csv"  # Replace with the path to your second CSV file
df2 = pd.read_csv(avg_std_dev)
```

```python
column_name = "Value"  # Replace with the actual column name from your CSV files
avg_ret_data = df1[column_name].values.tolist()
avg_std_dev_data = df2[column_name].values.tolist()
```

```python
bc_ant_avg_ret = 368.7731628417969
expert_ant_avg_ret = 10344.51757812
```

```python
x_values = np.arange(1, len(avg_ret_data)+ 1, step = 1)

# Create a plot
plt.figure(figsize=(10, 6))
plt.errorbar(x_values, avg_ret_data, yerr=avg_std_dev_data, fmt='o', capsize=5, label='Humanoid2d-v2 Average Returns and Average Std Devs')
plt.grid(True)
plt.plot(x_values, avg_ret_data)
plt.plot(x_values, [bc_ant_avg_ret for i in range(len(avg_ret_data))], label='Average BC Returns')
plt.plot(x_values, [expert_ant_avg_ret for i in range(len(avg_ret_data))], label='Expert Returns')
plt.xticks(x_values)

plt.legend()
plt.xlabel('Number of Itterations')
plt.ylabel('Average Returns with std deviations')
plt.title('Humanoid2d-v2 Average Returns and Average Std Devs vs Number of Itterations')
```
